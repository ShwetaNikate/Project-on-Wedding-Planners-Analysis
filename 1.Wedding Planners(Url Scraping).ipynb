```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import time

name_data = []
location_data = []
Planning_fee_data = []
type_demand_data = []
rating_data = []
review_data = []

pages = 20
print("Progress: ", end="")
cities = ["pune", "mumbai", "hyderabad"]
for city in cities:
    for i in range(1, pages+1):
        URL = f'https://www.wedmegood.com/vendors/{city}/wedding-planners/?page={i}'
        h = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/120.0.0.0 Safari/537.36"
    }
        response = requests.get(URL,headers=h)
        html_code = response.text
        soup = BeautifulSoup(html_code)

        for container in soup.find_all('div', attrs={'class': 'PrimaryVendorCard sc-jTzLTM wFnqD'}):
            # 1. 
            name = container.find('div', attrs={'class': 'frow'})
            name_data.append(name.text if name else np.nan)

            # 5. 
            location = container.find('div', attrs={'class': 'info-icon text-tertiary frow margin-r-10 protected-content jsEvent-onCopy'})
            location_data.append(location.text if location else np.nan)
    
             # 3. 
            Planning_fee = container.find('div', attrs={'class': 'frow f-wrap'})
            Planning_fee_data.append(Planning_fee.text if price else np.nan)

            # 6. 
            Type = container.find('div', attrs={'class': 'v-center margin-10 protected-content jsEvent-onCopy'})
            type_demand_data.append(Type.text if Type else np.nan)
            
            # 2. 
            rating = container.find('span', attrs={'class': 'StarRatingNew fw-500 center rating-new-5 regular'})
            rating_data.append(rating.text if rating else np.nan)
    
            # 4. 
            review = container.find('span', attrs={'class': 'review-cnt regular nowrap margin-l-5'})
            review_data.append(review.text if review else np.nan)
    
            
        
    
        time.sleep(2)
        print(".", end="")

print("\nN U M B E R 's")
print(f"{'Name':<25}: {len(name_data)}")
print(f"{'Location':<25}: {len(location_data)}")
print(f"{'Fee':<25}: {len(Planning_fee_data)}")
print(f"{'Type':<25}: {len(type_demand_data)}")
print(f"{'Ratings':<25}: {len(rating_data)}")
print(f"{'Review':<25}: {len(review_data)}")
```

```python
import pandas as pd

df = pd.DataFrame({"Name" : name_data,
                  "Location": location_data,
                   "Planning Fee": Planning_fee_data,
                   "Type & Demand": type_demand_data,
                   "Ratings": rating_data,
                   "Reviews": review_data})
df
```
